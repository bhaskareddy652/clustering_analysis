```python
X = df.select_dtypes(include=['float64', 'int64'])
X
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
scaler = StandardScaler()
df_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
model1= KMeans(n_clusters=3, random_state=101,n_init=1)
predicted1= model1.fit_predict(df_scaled)
scor1= silhouette_score(df_scaled, predicted1)
inertias = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=101,n_init=3)
    kmeans.fit(df_scaled)
    inertias.append(kmeans.inertia_)
plt.plot(range(1, 11), inertias, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
#used to find out the optimal numbeer of clusters
```

```python
from sklearn.cluster import AgglomerativeClustering
model2= AgglomerativeClustering(n_clusters=3)
predicted2= model2.fit_predict(df_scaled)
score2= silhouette_score(df_scaled, predicted2)
```

```python
from sklearn.cluster import DBSCAN
model3= DBSCAN(eps=0.5, min_samples=6)
predicted3= model3.fit_predict(df_scaled)
score3= silhouette_score(df_scaled, predicted3)
```

```python
from sklearn.mixture import GaussianMixture
model4 = GaussianMixture(n_components=5, random_state=101)
predicted4 = model4.fit_predict(df_scaled)
score4 = silhouette_score(df_scaled, predicted4)
logs= []
for i in range(1, 11):
    gm = GaussianMixture(n_components=i, random_state=101)
    gm.fit(df_scaled)
    logs.append(gm.score(df_scaled))  
plt.plot(range(1, 11),logs, marker='o')
plt.title('Elbow Method for Gaussian Mixture')
plt.xlabel('Number of Components')
plt.ylabel('Log Likelihood')
plt.show()
```

```python
print("silhouette Score for Kmeans:",scor1.round(2))
print("silhouette Score for Agglomerative Clustering:",score2.round(2))
print("silhouette Score for DBSCAN:",score3.round(2))
print("silhouette Score for GaussianMixture:",score4.round(2))
```

Agglomerative Clustering with Various Linkage methods-

```python
linkages = ['ward', 'complete', 'average', 'single']
scores = {}
for i in linkages:
    model2 = AgglomerativeClustering(n_clusters=4, linkage=i)
    predicted2 = model2.fit_predict(df_scaled)
    score2 = silhouette_score(df_scaled, predicted2)
    scores[i] = score2
    print(f"Silhouette Score for linkage='{i}':", score2.round(2))
plt.bar(scores.keys(), scores.values(), color='skyblue')
plt.title('Silhouette Scores for Different Linkage Methods')
plt.xlabel('Linkage Method')
plt.ylabel('Silhouette Score')
plt.show()
```
